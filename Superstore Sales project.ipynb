### UNIVARIATE ANALYSIS

Sales Distribution

```python
plt.figure(figsize=(7,4))
sns.histplot(df["Sales"], bins=30)
plt.title("Sales Distribution")
plt.show()
```

### Insight
- Sales distribution is right-skewed, indicating a few high-value transactions contribute significantly to total sales.

### Profit Distribution

Profit Distribution

```python
plt.figure(figsize=(7,4))
sns.histplot(df["Profit"], bins=30)
plt.title("Profit Distribution")
plt.show()
```

### Insight
- Profit includes negative values, showing loss-making transactions.

### CATEGORY-WISE ANALYSIS

Total Sales by Category

```python
category_sales = df.groupby("Category")["Sales"].sum()

plt.figure(figsize=(7,4))
category_sales.plot(kind="bar")
plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
```

### Insight
- Technology category contributes the highest sales.
- Furniture records the lowest sales.

### Total Profit by Category

```python
category_profit = df.groupby("Category")["Profit"].sum()

plt.figure(figsize=(7,4))
category_profit.plot(kind="bar")
plt.title("Total Profit by Category")
plt.xlabel("Category")
plt.ylabel("Profit")
plt.show()
```

### Insight
- Technology is the most profitable category.
- Furniture has lower profit margins.

### Sub-Category Analysis

```python
subcat_profit = df.groupby("Sub-Category")["Profit"].sum().sort_values()

plt.figure(figsize=(10,5))
subcat_profit.plot(kind="bar")
plt.title("Profit by Sub-Category")
plt.xticks(rotation=90)
plt.show()
```

### Insight
- Tables and Bookcases are loss-making sub-categories.
- Copiers and Phones generate high profits.

### Region-Wise Analysis

Sales by Region

```python
region_sales = df.groupby("Region")["Sales"].sum()

plt.figure(figsize=(7,4))
region_sales.plot(kind="bar")
plt.title("Sales by Region")
plt.show()
```

### Insight
- West region generates the highest sales.

### Profit by Region

```python
region_profit = df.groupby("Region")["Profit"].sum()

plt.figure(figsize=(7,4))
region_profit.plot(kind="bar")
plt.title("Profit by Region")
plt.show()
```

### Insight
- West is the most profitable region.
- Central region shows weaker profit performance.

### Discount and Profit Analysis

```python
plt.figure(figsize=(7,4))
plt.scatter(df["Discount"], df["Profit"], alpha=0.4)
plt.xlabel("Discount")
plt.ylabel("Profit")
plt.title("Discount vs Profit")
plt.show()
```

### Insight
- Higher discounts are associated with lower or negative profit.
- Excessive discounting impacts profitability.

### Sales and Profit Relationship

```python
plt.figure(figsize=(7,4))
plt.scatter(df["Sales"], df["Profit"], alpha=0.4)
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.title("Sales vs Profit")
plt.show()
```

### Insight
- Higher sales generally lead to higher profit, but not always due to discounts.

```python
sns.heatmap(df[['Sales','Quantity','Discount','Profit']].corr(), annot=True)
plt.title('Correlation Heatmap')
plt.show()
```

### Final Business Insights & Conclusion

- Technology category drives both sales and profit.
- Tables sub-category consistently incurs losses.
- High discounts negatively affect profit margins.
- West region outperforms other regions in sales and profit.
